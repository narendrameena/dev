```r

# paper pdf (local)#
#paper_pdf_local <- "/mnt/data/s41467-024-44994-w (4).pdf"

DefaultAssayName <- "SCT"
global_resolution <- 0.4

target_subcluster_counts <- list(
  Fibroblast = 11,
  Keratinocyte = 5,
  Myeloid = 11,
  Tcells = 6,
  Endothelial = 5,
  Pericyte = 2,
  Mast = 3,
  Melanocyte = 1,
  Other = 1
)
```

```r
# -------------------------
# 1) Load object
# -------------------------
combined <- readRDS(seurat_in)

cat("Loaded object with", ncol(combined), "cells and", nrow(combined), "genes.\n")

DefaultAssay(combined) <- DefaultAssayName

# Debug: Ensure seurat_clusters exists
if (!"seurat_clusters" %in% colnames(combined@meta.data)) {
  stop("ERROR: combined$seurat_clusters does NOT exist.")
}

combined$seurat_clusters <- as.character(combined$seurat_clusters)

cat("Unique clusters detected:", length(unique(combined$seurat_clusters)), "\n")

# Ensure UMAP exists
if (!"umap" %in% names(combined@reductions)) {
  cat("UMAP not found â€” computing now...\n")
  combined <- RunPCA(combined, verbose = FALSE)
  combined <- FindNeighbors(combined, dims = 1:30)
  combined <- RunUMAP(combined, dims = 1:30)
}
```
